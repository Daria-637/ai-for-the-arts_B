# From Print to Touch to AI: Reinterpreting the 1850 New Testament Through Accessibility and Generative AI

## Context

### Introduction

The bible is one of the most influential religious and cultural texts in human history. It is a collection of sacred writings central to Christianity, which includes teachings, narratives, poetry, and historical accounts that have shaped societies, moral values, literature, and art for centuries. The New Testament focuses on the life, teachings, death, and resurrection of Jesus Christ, as well as the development of early Christianity.

This project focuses on “The New Testament of Our Lord and Saviour Jesus Christ: Translated out of the Original Greek & with the Former Translations Diligently Compared & Revised”, published in 1850 by the American Bible Society. The American Bible Society is an organisation that has printed and distributed some 4 billion Bibles to date, and even in the 19th century played a key role in bringing religious texts to a wider audience. This book is particularly significant as it incorporated a format that allowed a wide range of readers, including those with visual disabilities, to read by touch through techniques such as embossing the text, prior to the widespread adoption of Braille. Therefore, this book is not only a religious text but also holds value as a cultural artefact illustrating the history of accessibility and inclusive reading practices.

In John 9:1–12, a man who had been blind from birth is depicted as a beggar, and Jesus’ disciples ask whether his blindness is due to his own sins or those of his parents. Jesus rejects this notion, replying instead that the man’s condition exists “so that God's works might be revealed in him” (Ridling, 1989, p. 2699). Like the story of the man healed by Jesus and restored to sight, this passage brings hope to people with visual disabilities. It is particularly significant in that it challenges conventional views of blindness and suffering. Rather than portraying blindness as a punishment, it presents the possibility of dignity, transformation, and new understanding. Therefore, this book is not only a religious text but also occupies an important place in the history of accessibility and inclusive reading practices.
